```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/refs/heads/master/titanic.csv"
df = pd.read_csv(url)
```

```python
num_cols= df.select_dtypes(include=['int64', 'float64']).columns
print(num_cols)
```

```python
corr_matrix = df[num_cols].corr()
print(corr_matrix)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
#heatmap makes it easier to spot strong or weak correlations.
```

Highest positive correlation:
Survived and Fare have a positive correlation (0.26).
People who paid higher fares had better survival chances.

Negative correlations:
Pclass and Fare have a strong negative correlation (-0.55).
Lower Pclass (i.e., 1st class)  Higher Fare paid.


ibSp and Parch have a positive correlation (0.41). If someone had siblings aboard, they were also more likely to have parents/children aboard.

Other variables show weak or no strong correlation.
